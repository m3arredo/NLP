```python
# New class for encoding that doesn't need a label column

class encoded_dataset1:

    def __init__(self, image_dir, df, tree, depth=8):
        """ Generate a tree based encoding for a set of images.
        image_dir: location of images
        df: dataframe with information about each image
        tree: the tree used for encoding
        depth: the number of tree levels to be used.
        """
        def bin2int(c):
            ans = 0
            for j in range(len(c)):
                bit = int(c[-(j+1)])
                ans += bit * (2 ** j)
            return ans
  
        self.df = df
        self.rows = df.shape[0]
        self.cols = 2 ** (depth + 1)
        data = zeros([self.rows, self.cols])

        j = 0
        self.codes = []
        for filename, row in df.iterrows():
            filepath = f"{image_dir}/{filename}"
            code = encode_image(filepath, tree)
            self.codes.append(code)
            V = zeros(self.cols)
            for c, a in code:
                V[bin2int(c)] = a

            data[j] = V

            if (j + 1) % 10 == 0:
                print(j, filename, end='\r')
            j += 1

        self.data = data

    def get_df(self):
        return self.df

    def get_slice(self, selection):
        assert selection.shape[0] == self.data.shape[0]
        S = self.data[selection, :]
        return S
```

# Random Forest
Using a random forest model to train on the train data and predict on the test.
I will play with the NN and RF and get the accuracies as high as possible, then use this to create my prediction with absention. Where the NN and RF agree, I go with that, where they disagree I go with 0. Which ever one has the higher accuracy is the one I use for prediction without abstention. 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Preprocess your data and split into features (X) and target labels (y)
X_train, X_val, y_train, y_val = train_test_split(X1, y, test_size=0.2, random_state=42)

# Instantiate the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

# Train the Random Forest classifier
rf_classifier.fit(X_train, y_train)

# Make predictions on the validation set
y_pred = rf_classifier.predict(X_val)
```
